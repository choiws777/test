# 데이터프레임 합치기

```python
df1 = pd.DataFrame({
    'a': ['a0', 'a1', 'a2'],
    'b': ['b0', 'b1', 'b2']}, index=[0, 1, 2])
 
df1
```

```python
df2 = pd.DataFrame({
    'a': ['a2', 'a3'],
    'b': ['b2', 'b3'],
    'c': ['c2', 'c3'],
    'd': ['d2', 'd3']}, index=[2, 3])

df2
```

### concat

```python
pd.concat([df1, df2], axis=0)
```

```python
pd.concat([df1, df2], axis=1)
```

```python
df_concat= pd.concat([df1, df2], axis=1)
```

```python
df_concat
```

### merge

```python
pd.merge(df1, df2, how="inner", on='b')
#교집합
```

```python
pd.merge(df1, df2, how="outer", on='b')
#합집합
```

```python
pd.merge(df1, df2, how="left", on='b')
#df_merge = pd.merge(df1, df2, how="left", on='b')
#df_merge
#df_merge.dropna() 결측치 제거
#df_merge.dropna(subset=['a_x'])
```

```python
pd.merge(df1, df2, how="right", on='b')
```

### join

```python
df1.join(df2, how='left', lsuffix='_l', rsuffix='_r')
```

```python
df1.join(df2, how='outer', lsuffix='_l', rsuffix='_r')
```

```python
df_pvt = pd.pivot_table(df,                        # 원본 데이터프레임
                     index='Gender',               # 행 위치
                     columns='Vehicle_Age',        # 열 위치
                     values='Annual_Premium',      # 데이터
                     aggfunc='sum')              # 집계 함수  sum  mean 등

df_pvt
```

```python
df_pvt[['< 1 Year', '1-2 Year', '> 2 Years']]
```

```python
df_pvt2 = pd.pivot_table(df,                        # 원본 데이터프레임
                     index=['Gender', 'Response'],               # 행 위치
                     columns='Vehicle_Age',        # 열 위치
                     values='Annual_Premium',      # 데이터
                     aggfunc=np.mean)              # 집계 함수

df_pvt2
```
