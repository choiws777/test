```python
review_sentence = ""
for review in data['tokens']:
    review_sentence = review_sentence + "" + review

print(review_sentence)
```

```python
total_words = okt.nouns(review_sentence)
len(total_words)
```

```python
# 한 글자 단어는 제거
final_words = [ word for word in total_words if len(word) > 1 ]   # List Comprehension 
print(len(final_words))
print(final_words[:10])
```

```python
# 단어 숫자 세기 
count = Counter(final_words)
tags = count.most_common(20)
print(tags)
```

```python
# 워드 클라우드 그리기
wordcloud = WordCloud(font_path='NanumBarunGothic', 
                      background_color='white', 
                      width=1200, 
                      height=800).generate_from_frequencies(dict(tags))

fig = plt.figure(figsize=(12, 8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
# 이미지 마스킹
from PIL import Image

image_path = '/content/drive/MyDrive/masking_WS.png'

icon=Image.open(image_path)
mask = Image.new("RGB", icon.size, (255,255,255))
mask.paste(icon, icon)
mask = np.array(mask)
```

```python
# 워드 클라우드 그리기
wordcloud = WordCloud(font_path='NanumBarunGothic', 
                      background_color='white', 
                      width=800, 
                      height=800, 
                      mask=mask).generate_from_frequencies(dict(tags))

fig = plt.figure(figsize=(10, 10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
